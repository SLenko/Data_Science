```python
import pandas as pd

# 1. Прочитайте файл 2017_jun_final.csv
df = pd.read_csv('2017_jun_final.csv')

# 2. Прочитайте отриману таблицю, використовуючи метод head
print(df.head())

# 3. Визначте розмір таблиці за допомогою методу shape
print(df.shape)

# 4. Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(df.dtypes)

# 5. Порахуйте, яка частка пропусків міститься в кожній колонці
missing_values = df.isnull().sum() / len(df)
print(missing_values)

# 6. Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
columns_to_drop = missing_values[missing_values > 0].index.tolist()
if 'Мова програмування' in columns_to_drop:
    columns_to_drop.remove('Мова програмування')
df_cleaned = df.drop(columns=columns_to_drop)

# 7. Знову порахуйте, яка частка пропусків міститься в кожній колонці
missing_values_after = df_cleaned.isnull().sum() / len(df_cleaned)
print(missing_values_after)

# 8. Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df_dropped = df.dropna()

# 9. Визначте новий розмір таблиці за допомогою методу shape
print(df_dropped.shape)

# 10. Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = df[df['Мова програмування'].str.contains('Python', na=False)]

# 11. Визначте розмір таблиці python_data за допомогою методу shape
print(python_data.shape)

# 12. Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped = python_data.groupby('Посада')

# 13. Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg
agg_data = grouped['Зарплата на місяць'].agg(['min', 'max']).reset_index()
print(agg_data)

# 14. Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць
def fill_avg_salary(row):
    return (row['min'] + row['max']) / 2

agg_data['avg'] = agg_data.apply(fill_avg_salary, axis=1)

# 15. Створіть описову статистику за допомогою методу describe для нового стовпчика
print(agg_data['avg'].describe())

# 16. Збережіть отриману таблицю в CSV файл
agg_data.to_csv('agg_data.csv', index=False)
```
